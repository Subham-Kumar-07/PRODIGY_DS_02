```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
sns.barplot(x='Sex', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Sex')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
sns.histplot(x='Age', hue='Survived', data=df, kde=True)
plt.title('Survival Rate by Age')
plt.show()
```

```python
sns.barplot(x='SibSp', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Sibling/Spouse Aboard')
plt.show()
```

```python
sns.barplot(x='Parch', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Parent/Children Aboard')
plt.show()
```

```python
sns.barplot(x='Embarked', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Embarked Port')
plt.show()
```
